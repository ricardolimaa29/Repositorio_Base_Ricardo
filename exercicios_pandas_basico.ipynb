## 🔹 1. Importar o Pandas
```python
import pandas as pd


---

## 🔹 2. Criar um DataFrame simples
```python
dados = {
    "Nome": ["Ana", "Bruno", "Carla"],
    "Idade": [23, 30, 27],
    "Cidade": ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]
}

df = pd.DataFrame(dados)
print(df)
```

---


## 🔹 3. Exibir informações gerais
```python
print(df.info())
```

---

## 🔹 4. Mostrar estatísticas básicas
```python
print(df.describe())
```

---

## 🔹 5. Selecionar uma coluna
```python
print(df["Idade"])
```

---

## 🔹 6. Selecionar várias colunas
```python
print(df[["Nome", "Cidade"]])
```

---

## 🔹 7. Filtrar linhas
```python
print(df[df["Idade"] > 25])
```

---

## 🔹 8. Adicionar uma nova coluna
```python
df["AnoNascimento"] = 2025 - df["Idade"]
print(df)
```

---

## 🔹 9. Ordenar os dados
```python
df_ordenado = df.sort_values(by="Idade", ascending=False)
print(df_ordenado)
```

---

## 🔹 10. Calcular média
```python
media = df["Idade"].mean()
print(f"Idade média: {media}")
```

---

## 🔹 11. Renomear colunas
```python
df = df.rename(columns={"Cidade": "Local"})
print(df)
```

---

## 🔹 12. Contar valores únicos
```python
print(df["Local"].value_counts())
```

---

## 🔹 13. Remover colunas
```python
df = df.drop(columns=["AnoNascimento"])
print(df)
```

---

## 🔹 14. Salvar em CSV
```python
df.to_csv("dados_novos.csv", index=False)
print("Arquivo salvo com sucesso!")
```
