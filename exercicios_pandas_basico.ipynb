## ðŸ”¹ 1. Importar o Pandas
```python
import pandas as pd


---

## ðŸ”¹ 2. Criar um DataFrame simples
```python
dados = {
    "Nome": ["Ana", "Bruno", "Carla"],
    "Idade": [23, 30, 27],
    "Cidade": ["SÃ£o Paulo", "Rio de Janeiro", "Belo Horizonte"]
}

df = pd.DataFrame(dados)
print(df)
```

---


## ðŸ”¹ 3. Exibir informaÃ§Ãµes gerais
```python
print(df.info())
```

---

## ðŸ”¹ 4. Mostrar estatÃ­sticas bÃ¡sicas
```python
print(df.describe())
```

---

## ðŸ”¹ 5. Selecionar uma coluna
```python
print(df["Idade"])
```

---

## ðŸ”¹ 6. Selecionar vÃ¡rias colunas
```python
print(df[["Nome", "Cidade"]])
```

---

## ðŸ”¹ 7. Filtrar linhas
```python
print(df[df["Idade"] > 25])
```

---

## ðŸ”¹ 8. Adicionar uma nova coluna
```python
df["AnoNascimento"] = 2025 - df["Idade"]
print(df)
```

---

## ðŸ”¹ 9. Ordenar os dados
```python
df_ordenado = df.sort_values(by="Idade", ascending=False)
print(df_ordenado)
```

---

## ðŸ”¹ 10. Calcular mÃ©dia
```python
media = df["Idade"].mean()
print(f"Idade mÃ©dia: {media}")
```

---

## ðŸ”¹ 11. Renomear colunas
```python
df = df.rename(columns={"Cidade": "Local"})
print(df)
```

---

## ðŸ”¹ 12. Contar valores Ãºnicos
```python
print(df["Local"].value_counts())
```

---

## ðŸ”¹ 13. Remover colunas
```python
df = df.drop(columns=["AnoNascimento"])
print(df)
```

---

## ðŸ”¹ 14. Salvar em CSV
```python
df.to_csv("dados_novos.csv", index=False)
print("Arquivo salvo com sucesso!")
```
